```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("../data/master/studytime2.csv")
df["day of week"] = pd.Categorical(df["day of week"], categories=["Sunday", "Monday", "Tuesday", "Wednesday", "Thursday", "Friday", "Saturday"], ordered=True)
```

```python
df
```

```python
df2 = df[df['term'] == 'sp24'].groupby(['day of week', 'course', 'term']).agg('mean').reset_index()[['day of week', 'course', 'duration']]
sns.barplot(df2, x='day of week', y='duration')
```

```python
df[df['term'] == 'fa24'][['week', 'day of week', 'duration']].groupby(["day of week", "week"]).agg('sum').reset_index()[['day of week', 'duration']].groupby('day of week').agg('mean')
```

```python
df[df['term'] == 'sp25'][['week', 'duration']].groupby(['week']).agg('sum') / 60
```

```python
df[df['term'] == 'sp25'][['week', 'course', 'duration']].groupby(['week', 'course']).agg('sum') / 60
```
